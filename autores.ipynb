```python
import numpy as np
import pandas as pd

import requests
from bs4 import BeautifulSoup

from PIL import Image
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt

import string
import nltk
from nltk.corpus import stopwords
```

```python
data = pd.read_csv("intelligent and assistant.csv")
```

```python
autores = data['Authors'].value_counts()
```

```python
plt.style.use('ggplot')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['axes.edgecolor'] = 'black'
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.color'] = 'grey'
```

```python
fig, ax = plt.subplots()
ax.barh(autores.head(8).index, autores.head(8).values, color = 'black')
ax.set_title('Cantidad de articulos por autor')
ax.set_xlabel('Autores')
ax.set_ylabel('Cantidad de articulos')
plt.show()
```
