```python
import numpy as np
import pandas as pd

import requests
from bs4 import BeautifulSoup

from PIL import Image
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt

import string
import nltk
from nltk.corpus import stopwords
```

```python
data = pd.read_csv("intelligent and assistant.csv")
```

```python
años = data['Year'].value_counts()
```

```python
años = años.sort_index()
```

```python
años
```

```python
plt.style.use('ggplot')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['axes.edgecolor'] = 'black'
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.color'] = 'grey'
```

```python
fig, ax = plt.subplots()
ax.plot(años.index, años.values, color = 'black')
ax.set_title('Artículos por año')
ax.set_xlabel('Año')
ax.set_ylabel('Artículos')
set_xticks = ax.set_xticks(años.index)
```
